```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns

image_path = '11.webp'

img = mpimg.imread(image_path)
if img.ndim == 3:
    img = img.mean(axis=2)

U, S, VT = np.linalg.svd(img, full_matrices=False)


correlation_columns = np.dot(VT.T, VT)
correlation_rows = np.dot(U, U.T)

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(correlation_rows, ax=axs[0], cmap='viridis')
axs[0].set_title("Macierz korelacji wierszy (XXᵀ)")

sns.heatmap(correlation_columns, ax=axs[1], cmap='viridis')
axs[1].set_title("Macierz korelacji kolumn (XᵀX)")

plt.tight_layout()
plt.show()
```
