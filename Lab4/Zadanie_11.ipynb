```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x):
    return x * sigmoid(x)

def swish_grad(x):
    sig = sigmoid(x)
    return sig + x * sig * (1 - sig)


x = np.linspace(-10, 10, 500)
y = swish(x)
dy = swish_grad(x)


plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Swish", linewidth=2)
plt.plot(x, dy, label="Swish - gradient", linestyle='--')
plt.title("Funkcja Swish i jej gradient")
plt.xlabel("x")
plt.ylabel("wartość")
plt.grid(True)
plt.legend()
plt.show()
```
